```python
def split_train_test(li,bckdays=22,fwddays = 5,is_expand=False):
    times = (len(li) - bckdays)//fwddays + 1
    trains = []
    tests = []
    for t in range(0,times):
        if is_expand:
            trains.append(li[:bckdays+t*fwddays])
        else:
            trains.append(li[t*fwddays:bckdays+t*fwddays])
        tests.append(li[bckdays+t*fwddays:bckdays+(t+1)*fwddays])
    return zip(trains,tests)
```

```python
li = range(0,20)
bckdays=5
fwddays = 2
times = (len(li) - bckdays)//fwddays + 1
trains = []
tests = []
for t in range(0,times):
    trains.append(li[t*fwddays:bckdays+t*fwddays])
    tests.append(li[bckdays+t*fwddays:bckdays+(t+1)*fwddays])
```
