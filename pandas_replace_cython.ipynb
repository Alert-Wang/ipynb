```python
def replace_value(df,val):
    for i in range(df.shape[0]):
        for j in range(df.shape[1]):
            if df.iat[i,j] > val:
                df.iat[i,j] = np.nan
```

```python
df = pd.DataFrame(np.random.randn(2500,2500))
%time replace_value(df,1.5)
```

```python
%%time
%%cython  
import numpy as np
cimport numpy as np
cimport cython

cdef int i = 0
cdef int j = 0

# @cython.cdivision(True)
# @cython.boundscheck(False)
# @cython.wraparound(False)
# @cython.profile(False)
cdef replace_value_cython(np.ndarray[np.float_t,ndim= 2] aa ,float val):
    for i in range(aa.shape[0]):
        for j in range(aa.shape[1]):
            if aa[i,j] > val:
                aa[i,j] = np.nan
aa = np.random.randn(2500,2500)
replace_value_cython(aa, 1.5)
```
